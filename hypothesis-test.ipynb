Quando ci sono tendenze nei dati, devo domandarmi se queste tendenze sono dovute a fluttuazioni casuali nella raccolta dei dati o a qualche fenomeno reale. Ed è qui che entra in gioco la verifica delle ipotesi. Diciamo che ho un set di dati fittizio che rappresenta un campione di alberi di avocado. Ora ci vogliono circa nove o 10 mesi perché un avocado cresca da un fiore a un frutto. Diciamo che ho un nuovo fertilizzante e voglio testare se il nuovo fertilizzante ha un effetto sul tempo di crescita dell'avocado. Quindi ho dei dati su un campione di alberi di avocado, e per ogni albero, il dataset contiene la quantità di tempo che l'avocado ha impiegato per crescere e se è stato usato il nuovo fertilizzante. Ora voglio determinare se c'è un'associazione tra la durata della crescita e se è stato usato il fertilizzante. Quindi progetterò un test d'ipotesi e creerò due ipotesi. 

# hypothesis test
Voglio capire se gli alberi di avocado crescono più velocemente se utilizzo del fertilizzante. 

Farà varie ipotesi:

## Null hypothesis
L'ipotesi nulla di solito non afferma che ci sono associazioni tra le variabili, e di solito attribuisce le tendenze osservate nel campione al caso. Quindi la mia ipotesi nulla sarà la seguente. Nella popolazione, la distribuzione della durata della crescita degli alberi di avocado che ricevono il fertilizzante è la stessa di quella degli alberi di avocado che non hanno ricevuto il fertilizzante. Se le due distribuzioni sono diverse nel campione, è dovuto al caso.

## Alternative hypothesis
Nella popolazione, la distribuzione della durata della crescita degli alberi di avocado che hanno ricevuto fertilizzante è diversa da quella degli alberi di avocado che non hanno ricevuto fertilizzante a causa di un'associazione tra la durata della crescita e se è stato utilizzato fertilizzante. E in media, gli alberi di avocado che ricevono fertilizzante, hanno impiegato un periodo di tempo più breve per crescere rispetto a quelli che non hanno ricevuto fertilizzante. 

## Obbiettivo di un test di ipotesi
Ora l'obbiettivo di un test di ipotesi è decidere tra l'ipotesi nulla e l'ipotesi alternativa.
